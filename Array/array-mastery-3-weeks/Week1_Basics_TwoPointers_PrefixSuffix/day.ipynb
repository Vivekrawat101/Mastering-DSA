```python
import os

base = "Day1_Basics_O_n"

file_content = {
    "code/01_brute_vs_smart.py": '''
# Problem: Find max subarray sum using brute and optimized approach.
# Input: [-2,1,-3,4,-1,2,1,-5,4]
# Output: 6
''',

    "code/02_find_max_min.py": '''
# Problem: Find max and min in the array.
# Input: [3, 7, 1, 9]
# Output: Max = 9, Min = 1
''',

    "code/03_prefix_sum.py": '''
# Problem: Compute prefix sum array.
# Input: [1, 2, 3]
# Output: [1, 3, 6]
''',

    "code/04_suffix_array.py": '''
# Problem: Compute suffix sum array.
# Input: [1, 2, 3]
# Output: [6, 5, 3]
''',

    "code/concepts_problems/max_diff_arr_i_j/problem.py": '''
# Problem: Find max (j - i) such that arr[i] < arr[j].
# Input: [1, 10]
# Output: 1
''',

    "code/concepts_problems/sum_even_index_subarrays/problem.py": '''
# Problem: Sum all subarrays starting from even indices.
# Input: [1, 2, 3]
# Output: 10
''',

    "code/concepts_problems/max_subarray_prefix_vs_kadane/problem.py": '''
# Problem: Max subarray sum using prefix sum vs Kadane.
# Input: [-2,1,-3,4,-1,2,1,-5,4]
# Output: 6
''',

    "reference/01_brute_vs_smart_ref.py": '''
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

# Brute-force
max_sum = float('-inf')
for i in range(len(arr)):
    for j in range(i, len(arr)):
        max_sum = max(max_sum, sum(arr[i:j+1]))
print("Brute:", max_sum)

# Kadane’s Algorithm
max_sum = curr = arr[0]
for num in arr[1:]:
    curr = max(num, curr + num)
    max_sum = max(max_sum, curr)
print("Kadane:", max_sum)
''',

    "reference/02_find_max_min_ref.py": '''
arr = [3, 7, 1, 9]

max_val = arr[0]
min_val = arr[0]

for num in arr:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Max:", max_val)
print("Min:", min_val)
''',

    "reference/03_prefix_sum_ref.py": '''
arr = [1, 2, 3]
prefix = []
curr = 0

for num in arr:
    curr += num
    prefix.append(curr)

print("Prefix Sum:", prefix)
''',

    "reference/04_suffix_array_ref.py": '''
arr = [1, 2, 3]
suffix = [0] * len(arr)
suffix[-1] = arr[-1]

for i in range(len(arr) - 2, -1, -1):
    suffix[i] = arr[i] + suffix[i + 1]

print("Suffix Array:", suffix)
''',

    "reference/concepts_problems/max_diff_arr_i_j/solution.py": '''
arr = [1, 10]
min_index = 0
max_diff = 0

for j in range(1, len(arr)):
    if arr[j] > arr[min_index]:
        max_diff = max(max_diff, j - min_index)
    if arr[j] < arr[min_index]:
        min_index = j

print(max_diff)
''',

    "reference/concepts_problems/sum_even_index_subarrays/solution.py": '''
arr = [1, 2, 3]
total = 0

for i in range(0, len(arr), 2):
    for j in range(i, len(arr)):
        total += sum(arr[i:j+1])

print(total)
''',

    "reference/concepts_problems/max_subarray_prefix_vs_kadane/solution.py": '''
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

# Kadane’s
max_sum = curr = arr[0]
for num in arr[1:]:
    curr = max(num, curr + num)
    max_sum = max(max_sum, curr)
print("Kadane:", max_sum)

# Prefix sum
prefix = [0] * len(arr)
prefix[0] = arr[0]
for i in range(1, len(arr)):
    prefix[i] = prefix[i-1] + arr[i]

min_prefix = 0
best = prefix[0]
for i in range(len(arr)):
    best = max(best, prefix[i] - min_prefix)
    min_prefix = min(min_prefix, prefix[i])

print("Prefix Sum:", best)
'''
}

# Create missing folders and write content
for rel_path, content in file_content.items():
    full_path = os.path.join(base, rel_path)
    os.makedirs(os.path.dirname(full_path), exist_ok=True)
    with open(full_path, "w", encoding="utf-8") as f:
        f.write(content.strip() + "\n")

print("✅ All Day 1 files created and filled successfully.")
```
