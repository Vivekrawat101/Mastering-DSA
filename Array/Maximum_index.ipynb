## ðŸ§® Problem: Maximum Index Difference (i < j and arr[i] < arr[j])

Given an array `arr` of positive integers, return the maximum value of `j - i` such that:

```
arr[i] < arr[j] and i < j
```

The goal is to find the **farthest possible pair of indices** where the value at the earlier index is strictly less than the value at the later index.

---

### âœ… Examples:

#### Example 1:
**Input:**
```python
arr = [1, 10]
```
**Output:**
```
1
```
**Explanation:**  
arr[0] = 1 and arr[1] = 10 â†’ 1 < 10 and 1 - 0 = 1.

---

#### Example 2:
**Input:**
```python
arr = [5, 4, 3]
```
**Output:**
```
0
```
**Explanation:**  
There is no pair of indices (i, j) where i < j and arr[i] < arr[j].

---

#### Example 3:
**Input:**
```python
arr = [34, 8, 10, 3, 2, 80, 30, 33, 1]
```
**Output:**
```
6
```
**Explanation:**  
The valid pair is (1, 7): arr[1] = 8 < 33 = arr[7]. The distance is 7 - 1 = 6.

---

### ðŸ”’ Constraints:

- `1 â‰¤ arr.size â‰¤ 10^5`
- `0 â‰¤ arr[i] â‰¤ 10^9`


````python
def maxIndexDiff(arr):
    
    n = len(arr)
    left_min = [0]*n
    right_max = [0]*n
    min = arr[0]
    max = arr[n-1]
    
    i=0
    
    while i < n:
        
        if arr[i] < min:
            min = arr[i]
        
        left_min[i] = min
            
        i+=1    
     
    
    
    
    i = n-1
    
    while i>=0:
        
        if arr[i] > max:
            max = arr[i]
        
        right_max[i] = max
        
        i-=1
    
    
    i=j=diff=0
    
    while i < n and j < n:
        
        if left_min[i] <= right_max[j]:
            
            if diff < j-i:
                diff = j-i
            j+=1
        else:
            i+=1
    print(diff)
    
    
    # print("arr      :", arr)
    # print("left_min :", left_min)
    # print("right_max:", right_max)
````
